# Creating a Supervised Learning Model:
Since we are using the features to identify the labels of the required result, weare doing a Supervised Learning.

We now split our data frame into train and test data, where the former one trains our learning model with some sample features and their labels.And after that, we use the latter part to test our trained model for accuracy of prediction.

# Classification model
When we see that the labels we need to predict are absolute values, and not continuous, we know we can't fit a line into it, and the picture of regression is out. So, for the yes-no result, we choose the classification model, binary trees.
In this, we have used Gini value to get the fit model.Further, we would use other values like entropy on binary trees, and also other methods of classification models,And compare the prediction accuracy.

```python
from sklearn.cross_validation import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
```

# Converting string values of the features into numeric
We can only train the binary tree classification model with numeric values and strings, and so we have to manually change the columns having string values into integer values.

```python
df1 = df
df1.Property_Area.replace(['Urban','Semiurban','Rural'],[2,1,0],inplace=True)
df1.Married.replace(['Yes','No'],[1,0],inplace=True)
df1.Gender.replace(['Male','Female'],[1,0],inplace=True)
df1.Education.replace(['Graduate','Not Graduate'],[1,0],inplace=True)
```

```python
df1.Self_Employed.replace(['Yes','No'],[1,0],inplace=True)
```

```python
df1
```

```python
df1.Dependents.replace('3+',3,inplace=True)
```

```python
x = df1.iloc[:,1:-1]
y = df1.iloc[:,-1]
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=.4,random_state=100)
```

```python
gini = DecisionTreeClassifier(criterion='gini',random_state=100,max_depth = 4,min_samples_leaf = 5)
gini.fit(x_train,y_train)
```

```python
y_predict = gini.predict(x_test,gini)
print(accuracy_score(y_test,y_predict))
```

The accuracy by using this model is 78.45 %
